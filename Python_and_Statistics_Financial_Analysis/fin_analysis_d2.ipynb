```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from path import Path
```

```python
fb = pd.read_csv(Path("resources/data/facebook.csv"))
fb
```

```python
fb["Price1"] = fb.Close.shift(-1)
fb
```

```python
fb["PriceDiff"] = fb.Price1 - fb.Close
fb
```

```python
fb["Returns_in_%"] = (fb.PriceDiff / fb.Close) * 100
fb
```

```python
fb["Direction"] = [1 if fb.loc[ei, "PriceDiff"]>0 else -1 for ei in fb.index]
fb["MA40"] = fb.Close.rolling(40).mean()
fb["MA200"] = fb.Close.rolling(200).mean()
```

```python
fb.Close.plot()
fb.MA40.plot()
fb.MA200.plot()
```

```python
msft = pd.read_csv(Path("resources/data/microsoft.csv"))
msft
```

```python
msft.Close.plot()
msft["MA10"] = msft.Close.rolling(40).mean().plot()
msft["MA50"] = msft.Close.rolling(200).mean().plot()
```

```python
msft["Shares"] = [1 if msft.loc[ei, "MA10"] > msft.loc[ei,"MA50"] else 0 for ei in msft.index]
msft
```
